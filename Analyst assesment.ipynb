```python
import pandas as pd

```

```python
df = pd.read_csv('C:/Data Analyst/funnel_events_sample.csv')
```

```python
funnel_order = ['visited_site', 'signup_started', 'details_filled', 'email_verified', 'purchase_completed']
```

```python
funnel = df.groupby('step')['user_id'].nunique().reindex(funnel_order).reset_index()
funnel.columns = ['step', 'unique_users']
funnel['prev_users'] = funnel['unique_users'].shift(1)
funnel['conversion_rate'] = (funnel['unique_users'] / funnel['prev_users'] * 100).round(2)
funnel['drop_off'] = funnel['prev_users'] - funnel['unique_users']
```
